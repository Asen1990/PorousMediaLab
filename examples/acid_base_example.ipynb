```python
import matplotlib.pyplot as plt
import sys
sys.path.append('../src')
from PorousMediaLab import PorousMediaLab
import numpy as np
%matplotlib inline
```

```python
# tend = 10
tend = 2
dt = 0.003
dx = 0.1 
L = 10
x = np.linspace(0, L, L / dx + 1)
t = np.linspace(0, tend, round(tend / dt) + 1)
phi =  0.9 
```

```python
ftc = PorousMediaLab(L, dx, tend, dt, phi)
```

```python
C_init=0.0000001/phi/3
ftc.add_species(is_solute=True, element='H2CO3', D=4, init_C=C_init, bc_top=1e-3, bc_top_type='flux', bc_bot=0, bc_bot_type='dirichlet')
ftc.add_species(is_solute=True, element='HCO3', D=4, init_C=C_init, bc_top=1e-3, bc_top_type='flux', bc_bot=0, bc_bot_type='dirichlet')
ftc.add_species(is_solute=True, element='CO3', D=4, init_C=C_init, bc_top=1e-3, bc_top_type='flux', bc_bot=0, bc_bot_type='dirichlet')

ftc.add_acid(species=['H2CO3', 'HCO3', 'CO3'], pKa=[3.6, 10.32])


ftc.add_species(is_solute=True, element='H2SO4', D=4, init_C=C_init,  bc_top=1e-3, bc_top_type='flux', bc_bot=0, bc_bot_type='dirichlet')
ftc.add_species(is_solute=True, element='HSO4',  D=4, init_C=C_init,  bc_top=1e-3, bc_top_type='flux', bc_bot=0, bc_bot_type='dirichlet')
ftc.add_species(is_solute=True, element='SO4',  D=4, init_C=C_init,  bc_top=1e-3, bc_top_type='flux', bc_bot=0, bc_bot_type='dirichlet')

ftc.add_acid(species=['H2SO4', 'HSO4', 'SO4'], pKa=[-10, 1.99])

ftc.add_species(is_solute=True, element='Ca', D=4, init_C=1e-9, bc_top=0., bc_top_type='dirichlet', bc_bot=1e-3, bc_bot_type='flux')
ftc.add_species(is_solute=False, element='CaCO3', D=1e-2, init_C=0., bc_top=0.0, bc_top_type='flux', bc_bot=0.00, bc_bot_type='flux')
ftc.add_species(is_solute=False, element='CaSO4', D=1e-2, init_C=0., bc_top=0.0, bc_top_type='flux', bc_bot=0.00, bc_bot_type='flux')


ftc.add_ion(element='Ca', charge=2)
```

```python
ftc.constants['Ks_CaCO3'] = 8e-9
ftc.constants['Ks_CaSO4'] = 2.4e-9
ftc.constants['k_pre'] = 1e-2
ftc.constants['CF'] = (1-phi)/phi  ### conversion factor

```

```python
ftc.rates['R_pre_CaCO3'] = 'k_pre * tanh(Ca*CO3/Ks_CaCO3-1)'
ftc.rates['R_pre_CaSO4'] = 'k_pre * tanh(Ca*SO4/Ks_CaSO4-1)'
```

```python
ftc.dcdt['CaCO3'] = 'R_pre_CaCO3/CF'
ftc.dcdt['Ca'] = '-R_pre_CaCO3*CF  - R_pre_CaSO4*CF'
ftc.dcdt['CO3'] = '-R_pre_CaCO3*CF'
ftc.dcdt['CaSO4'] = 'R_pre_CaSO4/CF'
ftc.dcdt['SO4'] = '-R_pre_CaSO4*CF'
```

```python
# %%timeit
ftc.solve()
```

```python
ftc.plot_contourplots(labels=False)
```

```python
ftc.plot_profiles()
```

```python
plt.plot(ftc.profiles.CaCO3, -ftc.x , label='$CaCO_3$', lw=4)
plt.plot(ftc.profiles.CaSO4, -ftc.x , label='$CaSO_4$', lw=4)
plt.xlim(-.0,0.05)
plt.ylim(-10,0.0)
# plt.axvline(0, c='k', ls='-.')
plt.ylabel('Depth')
plt.xlabel('Concnetration')
plt.legend(frameon=1)
```
