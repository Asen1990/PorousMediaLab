```python
ftc.add_species(is_solute=True, element='SF6w', D=D_SF6*theta, init_C=SF6_init, bc_top=0., bc_top_type='dirichlet', bc_bot=0, bc_bot_type='flux')
ftc.add_species(is_solute=True, element='SF6a', D=D_SF6air*(1-theta), init_C=0, bc_top=0., bc_top_type='dirichlet', bc_bot=0, bc_bot_type='flux')
```

```python
for i in range(1, len(ftc.time)):
    ftc.integrate_one_timestep(i)
    Fx[i] = ftc.estimate_flux_at_top('SF6a', i)
    if (ftc.time[i] > Tm[0] and ftc.time[i] < Tm[1]):
        Chs[i] = Chs[i-1] +  Fx[i] * SA / Vh * dt
    else:
        Chs[i] = 0
    ftc.SF6a.bc_top = 0
```
